## AutoML Training In Azure Machine Learning Service

This tutorial will go over how to train a model on the [IBM employee attrition dataset](data.csv) using AutoML on Azure Machine Learning service.

Go to the [Azure ML Studio](https://ml.azure.com) and click on the **Automated ML**. Then click on **+ New Automated ML run**.

![1](screenshots/1.png)

Click **+ Create dataset** and then select **From local files**. Then click **Next**.

![2](screenshots/2.png)

Type **IBM-Employee-Attrition** as the name of the dataset. Then click **Next**. 

![1](screenshots/3.png)

Upload the dataset csv file and change the upload path to **ibm-employee-attrition**. Then click **Next**. 

![1](screenshots/4.png)

Change the **Column headers** selection to **Use headers from the first file**. Keep everything else the same and click next.

![1](screenshots/5.png)

Select the columns you want to consider in the training process. Then click **Next**.

![1](screenshots/6.png)

Confirm that your summary matches the summary below. Then click **Create**. 

![1](screenshots/7.png)

Once your dataset has been created, you can now select it from the list. Make sure to have the dataset selected, and then click **Next**.

![1](screenshots/8.png)

Name the experiment **ibm-automl** and specify **Attrition** as the target column. This is the column you would like the model to try and predict. 

If you already have a compute target that you would like to use, select that target (and skip the next step). If not, click on **Create a new compute**.

![1](screenshots/9.png)

Create your compute target, following the configurations below. Then click **Create**, followed by **Next**.

![1](screenshots/10.png)

Click on **View additinoal configuration settings**.

![11](screenshots/11.png)

Expand the **Exit criterion** dropdown. Set **training job time** to 1 hour and **Metric score threshold** to 0.88. This will complete the automl run once a model with 88%+ accuracy is found or 1 hour has elapsed, whichever comes first. Click **Save**, followed by **Finish**. 

Your run should now be starting.

![12](screenshots/12.png)

While the AutoML run is executing, you can click on the **Models** tab to see all the child runs.

![13](screenshots/13.png)

As shown below, you can see all the child runs that the AutoML has automatically triggered along with it performance.

![14](screenshots/14.png)
